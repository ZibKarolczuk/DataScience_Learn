```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("./ml-1m/movies.dat", 
                     sep="::", engine="python", 
                     header=None, 
                     names=["MovieID","Title","Genres"])
```

```python
users = pd.read_csv("./ml-1m/users.dat", 
                    sep="::", engine="python", 
                    header=None, 
                    names=["UserID","Gender","Age","Occupation","Zip-code"])
```

```python
ratings = pd.read_csv("./ml-1m/ratings.dat", 
                      sep="::", engine="python", 
                      header=None,
                      names=["UserID","MovieID","Rating","Timestamp"])
```

```python
#Rok produkcji filmu
```

```python
# movies.Title.str[-5:-1]
# movies

movies["Year"] = movies.Title.str.extract(r"(\d{4})\)$")[0]
```

```python
movies.Title = movies.Title.str[:-6]
movies
```

```python
movies.Year = movies.Year.astype(int)
```

```python
movie_rating = pd.DataFrame(pd.merge(movies, ratings))
movie_rating

pd.pivot_table(movie_rating[["Title", "Rating"]],
               index = ["Title"],
               aggfunc = "mean"
              ).sort_values(by=["Rating", "Title"], ascending=[False, True])
```

```python
pd.pivot_table(movie_rating[["Title", "Rating"]],
               index = ["Title"],
               aggfunc = ["mean", "size"],
              )#sort_values(by=[("Rating","mean"), "Title"], ascending=[False, True])
```

```python
srednieOceny = ratings.groupby("MovieID").agg({"Rating":["mean", "size"]})
srednieOceny100 = srednieOceny[srednieOceny[("Rating", "size")] > 100]
srednieOceny100.columns = ["mean", "size"]
pd.merge(srednieOceny100, movies, left_index=True, right_on="MovieID")\
            .sort_values(by="mean", ascending = False)
```

```python
rating_users = pd.merge(ratings, users)
rating_users.head()

srednieOcenyPlcie = rating_users.groupby(["MovieID", "Gender"]).agg({"Rating":["mean", "size"]})
# srednieOceny100 = srednieOceny[srednieOceny[("Rating", "size")] > 100]

srednieOcenyPlcie.columns = ["mean", "size"]
srednieOcenyPlcie = srednieOcenyPlcie.reset_index()
srednieOcenyPlcie100 = srednieOcenyPlcie[srednieOcenyPlcie["size"] > 100]
pd.merge(srednieOcenyPlcie100, movies)\
        .sort_values(by="mean", ascending=False)\
        .drop_duplicates(subset="Gender")
```
