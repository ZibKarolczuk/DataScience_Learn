```python
sql_result3 = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result3
```

```python
#TABELA KONTYNGENCJI
wynik = planes.engine.value_counts()
# wynik.reset_index
# print(wynik).sort(ascending = True)
wynik = wynik.reset_index()[["engine", "index"]].sort_values(by="index").reset_index(drop=True)
wynik = wynik.rename({"engine" : "COUNT(*)", "Index" : "engine"}, axis=1)
wynik
```

```python
#TABELA PRZESTAWNA
wynik_pivot = pd.pivot_table(planes[["engine"]],
                             index = ["engine"],
#                              columns = ["index"],
                             aggfunc = "size"
                            )
wynik_pivot = wynik_pivot.reset_index().rename({"index" : "engine", "0" : "COUNT(*)"}, axis=1)
wynik_pivot
```

```python
#GROUP BY
# planes.groupby("engine", as_index=False).count() #zlicza tylko warto≈õc nie NA
pandas_wynik3groups = planes.groupby("engine", as_index=False).size() #zlicza wszystkie wiersze
pandas_wynik3groups.reset_index()
```

```python
sql_result4 = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type", con)
sql_result4
```

```python
pd.crosstab(planes.engine, planes.type).stack()
# pd.crosstab(planes.engine, planes.type)
# pd.crosstab(planes.engine, planes.type, margins=True).reset_index()#.rename({"All" : "COUNT(*)"}, axis = 1)[["type", "engine"]]
```
