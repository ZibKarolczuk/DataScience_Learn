```python
teraz = datetime.datetime.now()
t = np.linspace(0, 2*np.pi)
x0 = y0 = 0
r = 1

x = x0 + r * np.cos(t)
y = x0 + r * np.sin(t)

def angleHr(godzina):
    return ( (np.pi/2) - ((godzina/12)*2*np.pi) )
def angleMn(godzina):
    return ( (np.pi/2) - ((godzina/60)*2*np.pi) )

godzina = 12
minuta = 29

xh = x0 + 0.4 *np.cos( angleHr(godzina) )
yh = y0 + 0.4 *np.sin( angleHr(godzina) )

xm = x0 + 0.9 *np.cos( angleMn(godzina) )
ym = y0 + 0.9 *np.sin( angleMn(godzina) )

plt.fill(x,y, edgecolor = "red", linewidth=2)
plt.plot([0,xh],[0,yh], 'k-', alpha=0.95)
plt.plot([0,xm],[0,ym], 'k-', alpha=0.95)

plt.axis("equal")
plt.show()
teraz
```
